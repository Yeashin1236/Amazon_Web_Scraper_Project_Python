```python
# Import Libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
# Connect to the website and pull in the data
url = "https://www.amazon.com/Funny-Data-Systems-Business-Analyst/dp/B07FNW9FGJ"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36",
           "Accept-Language": "en-US,en;q=0.9",
           }

#URL for headers in own pc (https://httpbin.org/get)
# Make the request
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
print(soup)
```

```python
# Extract product title
title = soup.find(id="productTitle")
if title:
    print(title.get_text(strip=True))
```

```python
# Extract product rating
rating = soup.find("span", {"class": "a-icon-alt"})
if rating:
    print(rating.get_text(strip=True))
    
```

```python
def check_rating():
    url = "https://www.amazon.com/Funny-Data-Systems-Business-Analyst/dp/B07FNW9FGJ"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, "html.parser")
        
        title = soup.find(id="productTitle").get_text().strip()
        total_rating = soup.find(id="acrCustomerReviewText").get_text().strip()
        rating = soup.find("span", {"class": "a-icon-alt"}).get_text().strip()
        
        return [title, rating, total_rating, str(datetime.date.today())]
    
    except Exception as e:
        print(f"Error scraping data: {e}")
        return None

# File path
file_path = r"F:\Python project\Amazon Web Scraper Project 1\Amazonwebscraping.csv"
header = ['Product Title', 'Rating', 'Total_Rating', 'Date']

try:
    while True:
        data = check_rating()
        if data:
            with open(file_path, 'a', newline='', encoding='UTF8') as f:
                writer = csv.writer(f)
                if f.tell() == 0:  # Write header only if file is empty
                    writer.writerow(header)
                writer.writerow(data)
            print(f"Data saved: {data}")
        
        print("Waiting 5 seconds... (Press Ctrl+C to stop)")
        time.sleep(5)

except KeyboardInterrupt:
    print("\nScraping stopped by user")
except Exception as e:
    print(f"Error: {e}")
```
